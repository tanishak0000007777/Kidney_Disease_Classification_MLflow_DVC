```
@ensure_annotations

def get_predict(x:int ,y:int )->int:
    return x*y
```

 ## How to fetch data from google drive


```
import gdown
```

```
url='https://drive.google.com/file/d/1zc-wA-zeQ_EEblpoJfOCX6HlOeomsbaW/view?usp=sharing'
```

```
# see we have to get the unique id from the link

file_id=url.split("/")[-2]
```

```
file_id
```

```
prefix='https://drive.google.com/uc?/export=dwonload&id='
```

```
gdown.download(prefix+file_id,"kidney-CT-Scan_data.zip")
```
